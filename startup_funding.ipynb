```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv("/content/startup_funding.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df['Investors Name'] = df['Investors Name'].fillna('Undisclosed')
```

```python
df['Investors Name']
```

```python
sorted(df['Investors Name'].unique().tolist())
```

```python
df.head()
```

```python
df.drop(columns=['Remarks'],inplace=True)
```

```python
df.head()
```

```python
df.set_index('Sr No',inplace=True)
```

```python
df
```

```python
df.rename(columns={
    'Date dd/mm/yyyy':'Date',
    'Startup Name':'startup',
    'Industry Vertical':'Vertical',
    'SubVertical':'SubVertical',
    'City  Location':'City',
    'Investors Name':'investors',
    'InvestmentnType':'InvestmentType',
    'Amount in USD':'Amount'

},inplace=True)
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
df['Amount'] = df['Amount'].fillna(0)
```

```python
df.info()
```

```python
df['Amount'] = (
    df['Amount']
    .astype(str)
    .str.replace(',', '', regex=False)
    .str.replace('undisclosed', '0', regex=False)
)
```

```python
df['Amount'] = (
    df['Amount']
    .astype(str)
    .str.replace(',', '', regex=False)   # remove commas
    .str.replace('undisclosed', '0', regex=False)
    .pipe(pd.to_numeric, errors='coerce')
    .fillna(0)
    .astype(int)
)
```

```python
df['Amount']
```

```python
df.info()
```

```python
df.head()
```

```python
def to_inr(dollar):
  inr = dollar * 82.5
  return inr/10000000
```

```python
df['Amount'] = df['Amount'].apply(to_inr)
```

```python
df['Amount']
```

```python
df.head()
```

```python
df['Date'].replace()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='mixed', dayfirst=True, errors='coerce')
```

```python
df['Date']
```

```python
df.head()
```

```python
df.info()
```

```python
df= df.dropna(subset=['Date','startup','Vertical','City','investors' ,'InvestmentType' ,'Amount'])
```

```python
df.info()
```

```python
df.to_csv('Startup_cleaned.csv', index=False)
```
