```python
import nlpcloud


# NLP APPLICATION CLASS
class NLPApp:

    # CONSTRUCTOR
    def __init__(self):
        self.__database = {}
        self.__first_menu()   # call first menu

    # FIRST MENU
    def __first_menu(self):
        first_input = int(input(
            """\nHey, how would you like to proceed?
1. Not a member? Register
2. Already a member? Login
3. Exit
"""
        ))

        if first_input == 1:
            self.__register()
        elif first_input == 2:
            self.__login()
        else:
            print("Thank you!")
            exit()

    # REGISTER
    def __register(self):
        print("\n--- Register ---")
        name = input("Enter name: ")
        email = input("Enter email: ")
        password = input("Enter password: ")

        if email in self.__database:
            print("Email already exists!")
        else:
            self.__database[email] = [name, password]
            print("Registration successful!")

        self.__first_menu()

    # LOGIN
    def __login(self):
        print("\n--- Login ---")
        email = input("Enter email: ")
        password = input("Enter password: ")

        if email in self.__database:
            if password == self.__database[email][1]:
                print("Login successful!")
                self.__second_menu()
            else:
                print("Incorrect password!")
                self.__login()
        else:
            print("Email not registered!")
            self.__first_menu()

    # SECOND MENU
    def __second_menu(self):
        second_input = int(input(
            """\nWelcome! Choose an option:
1. Named Entity Recognition (NER)
2. Language Detection
3. Sentiment Analysis
4. Logout
"""
        ))

        if second_input == 1:
            self.__ner()
        elif second_input == 2:
            self.__language_detection()
        elif second_input == 3:
            self.__sentiment_analysis()
        else:
            print("Logged out successfully!")
            self.__first_menu()

    # NAMED ENTITY RECOGNITION
    def __ner(self):
        para = input('Enter the paragraph: ')
        search_term = input('What would you like to search: ')

        client = nlpcloud.Client(
            "gpt-oss-120b",
            "47a18fb33bfd632f0784fbfe3c5e0482253b3cdd",
            gpu=True
        )

        response = client.entities(
            para,
            searched_entity=search_term
        )

        print(response)
        self.__second_menu()

    # LANGUAGE DETECTION
    def __language_detection(self):
        print("Language Detection functionality will be added here")
        self.__second_menu()

    # SENTIMENT ANALYSIS
    def __sentiment_analysis(self):
        para = input('Enter the paragraph: ')

        client = nlpcloud.Client(
            "gpt-oss-120b",
            "47a18fb33bfd632f0784fbfe3c5e0482253b3cdd",
            gpu=False
        )

        response = client.sentiment(para)

        l=[]
        for i in response['secored_labels']
           l.append(i['score'])

        index= sorted(list(enumerate(l)),key=lambda x:x[1],reverse=True)[0][0]

        print(response['scored_labels'][index]['labels'])
        self.__second_menu() ## if we do not do it then it again start from the strting firest menu se



# OBJECT CREATION
app = NLPApp()
```
