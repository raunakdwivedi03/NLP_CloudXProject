```python
class NLPApp:

    def __init__(self):
        self.__database = {}
        self.__first_menu()   # call first menu

    def __first_menu(self):
        first_input = int(input(
            """\nHey, how would you like to proceed?
1. Not a member? Register
2. Already a member? Login
3. Exit
"""
        ))

        if first_input == 1:
            self.__register()
        elif first_input == 2:
            self.__login()
        else:
            print("Thank you!")
            exit()

    def __register(self):
        print("\n--- Register ---")
        name = input("Enter name: ")
        email = input("Enter email: ")
        password = input("Enter password: ")

        if email in self.__database:
            print("Email already exists!")
        else:
            # dictionary structure: email -> [name, password]
            self.__database[email] = [name, password] ## this is the dictornary
            print("Registration successful!")

        self.__first_menu()

    def __login(self):
        print("\n--- Login ---")
        email = input("Enter email: ")
        password = input("Enter password: ")

        if email in self.__database:
            if password == self.__database[email][1]:
                print("Login successful!")
                self.__second_menu()
            else:
                print("Incorrect password!")
                self.__login()
        else:
            print("Email not registered!")
            self.__first_menu()

    def __second_menu(self):
        second_input = int(input(
            """\nWelcome! Choose an option:
1. Named Entity Recognition (NER)
2. Language Detection
3. Sentiment Analysis
4. Logout
"""
        ))

        if second_input == 1:
            print("NER functionality will be added here")
            self.__second_menu()

        elif second_input == 2:
            print("Language Detection functionality will be added here")
            self.__second_menu()

        elif second_input == 3:
            print("Sentiment Analysis functionality will be added here")
            self.__second_menu()

        else:
            print("Logged out successfully!")
            self.__first_menu()


# object creation
app = NLPApp()
```
